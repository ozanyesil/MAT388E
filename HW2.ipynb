I created a function `transpose` which return the transpose of the pandas Series which makes easier to interpret.

```python
def transpose(dfr):
    return np.transpose(dfr)
```

```python
data = pd.DataFrame()
data['lt'] = literacy['TUR']
data['lf'] = labor_force['TUR']
data['pv'] = poverty['TUR']
data['hlt'] = health['TUR']
data['gdp'] = gdp['TUR']
data['mrt'] = mortality['TUR']
```

**1. Write a function that takes a ticker symbol and returns a pandas dataframe that for each day puts a 1 when the closing price is higher than the opening price, a 0 when the closing price is lower than the opening price.**

I wrote a function named `ticker`. It takes the ticker symbol as the input and creates a list that gives 1 or 0 depending pn the comparison pf closing and opening price. Then, it adds that list as the new column to the dataframe and returns it. I named the new column as `Close > Open`.

```python
def ticker(symbol):
    l = []
    yahoo_df = yf.download(str(symbol))
    for i in range(len(yahoo_df)):
        if(yahoo_df.iloc[i]['Close'] > yahoo_df.iloc[i]['Open']):
            l.append(1)
        else:
            l.append(0)
    yahoo_df['Close > Open'] = l
    return yahoo_df
```

For this question use the following [data](https://archive.ics.uci.edu/ml/datasets/credit+approval):


```python
credit = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/credit-screening/crx.data', header=None)

fn = {'+': 1, '-': 0}

X = credit.replace('?',0).iloc[:,[1,2,7,10,14]]
y = credit.iloc[:,15].map(lambda x: fn.get(x,0))
```

1. Split the data into training and test set.
2. Construct a logistic regression model predicting y against X.
3. Score the model.
4. Construct a [confusion matrix](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html) on the test data set and evaluate the model.
5. Analyse your results from Steps 3 and 4.

**1. Split the data into training and test set.**

I splitted data with function of sklearn.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

**2. Construct a logistic regression model predicting y against X.**

I assigned LogisticRegression. to `model`.  

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
model = LogisticRegression(max_iter=1500)
model.fit(X_train,y_train)
```

**3. Score the model.**

Scoring model with **model.score**

```python
print(model.score(X_test,y_test))
```

**4. Construct a [confusion matrix](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html) on the test data set and evaluate the model.**

I constructed it with the function `confusion_matrix`

```python
y_predict = model.predict(X_test)
confusion_matrix(y_test,y_predict)
```
